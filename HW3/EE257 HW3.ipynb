```python
# Split data into training / test
import numpy as np
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.20, random_state=42)

# data for 5 or not 5
# modify target to 5 or not 5
for i in range(len(five_ornot.target)):
    if five_ornot.target[i] != 5:
        five_ornot.target[i] = 0

five_train, five_test, five_ytrain, five_ytest = train_test_split(five_ornot.data, five_ornot.target, test_size=0.20, random_state=42)
```

##### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = KNeighborsClassifier(n_neighbors=5)
neighbors.fit(five_train, five_ytrain)

# Return average accuracy on given test data and labels
neighbors.score(five_test, five_ytest)
```

#### Logistic Regression

```python
import sklearn.linear_model as skl_lm

# Using newton-cg solver, the coefficients are equal/closest to the ones in the book. 
clf = skl_lm.LogisticRegression(solver='newton-cg')
clf.fit(five_train.values, five_ytrain)
print(clf)
print('classes: ',clf.classes_)
print('coefficients: ',clf.coef_)
print('intercept :', clf.intercept_)
```

```python
# Return mean accuracy on given test data and labels
clf.score(five_test.values , five_ytest)
```
